```python
#importing required library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#importing the data set
df = pd.read_csv('Virat_Kohli_Performance_2008_2025_Synthetic.csv')
```

```python
df.head()
```

```python
#visualizing the data by usin the subplots
fig, ax = plt.subplots(1, 2, figsize=(18, 5))
# 1st Subplot: Bar Plot (Runs Over Years)
ax[0].bar(df['Year'], df['Runs'])
ax[0].set_title("Virat Kohli IPL Runs (2008–2025)")
ax[0].set_xlabel("Year")
ax[0].set_ylabel("Runs")
 # 2nd Subplot: Line Plot (Batting Average Over Years)
ax[1].plot(df['Year'], df['Average'] , marker='*')
ax[1].set_title("Virat Kohli IPL Batting Average (2008–2025)")
ax[1].set_xlabel("Year")
ax[1].set_ylabel("Batting Average")
plt.tight_layout()
plt.show()
```
