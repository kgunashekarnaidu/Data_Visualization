<center><h1>Data Visualization using Seaborn Librabry</h1></center>

```python
#Importing required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the required Dataset
df = pd.read_csv('sales_data.csv')
```

```python
#Understating the dataset
df.head()
```

```python
#information of the dataset
df.info()
```

```python
#finding the shape of the data set
df.shape
```

<h3>Plot a scatterplot to analyze the relationship between Unit price and customer Rating Based on Gender</h3>

```python
#scatterplot
plt.figure(figsize=(8,6))
sns.scatterplot(data=df,x='Unit_Price',y='Customer_Rating',hue='Gender')
plt.title('Unit Price vs Customer Rating Based on Gender')
plt.show()
```

<h3>Plot the distribution of Sales_Amount with histogram and KDE curve</h3>

```python
#Histogram 
plt.figure(figsize=(7,5))
sns.histplot(df['Sales_Amount'], )
plt.title('Distribution of Sales Amount')
plt.xlabel('Sales Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
#Histogram and KDE Curve
plt.figure(figsize=(7,5))
sns.histplot(df['Sales_Amount'], kde=True)
plt.title('Distribution of Sales Amount')
plt.xlabel('Sales Amount')
plt.ylabel('Frequency')
plt.show()
```

<h3>Show the number of purchases made in each Region with a count plot</h3>

```python
#Count Plot
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Region')
plt.title('Number of Purchases made by Region')
plt.show()
```

<h3>Plot the distribution of Sales_Amount for each Product</h3>

```python
#Box Plot
plt.figure(figsize=(8,6))
sns.boxplot(data=df, x='Product', y='Sales_Amount')
plt.title('Sales Amount Distribution by Product')
plt.show()
```
